# Normalizing Data

```python
def normalize_arabic(text):
    text = re.sub("[إأآا]", "ا", text)
    text = re.sub("ى", "ي", text)
    text = re.sub("ؤ", "و", text)
    text = re.sub("ئ", "ي", text)
    text = re.sub("ة", "ه", text)
    text = re.sub("[^ء-ي\s]", "", text)
    text = re.sub(r'(.)\1+', r"\1\1", text) # removing longation
    text = re.sub(r'\s+', ' ', text).strip() # removing spaces
    text = re.sub(r'[^\u0600-\u06FF\s]', '', text) # removing non-Arabic characters
    return text
```

# Removing Diacritics

```python
def remove_diacritics(text):
    diacritics = re.compile("""
                             ّ    | # Shadda
                             َ    | # Fatha
                             ً    | # Tanwin Fath
                             ُ    | # Damma
                             ٌ    | # Tanwin Damm
                             ِ    | # Kasra
                             ٍ    | # Tanwin Kasr
                             ْ    | # Sukun
                             ـ     # Tatwil/Kashida
                         """, re.VERBOSE)
    return re.sub(diacritics, '', text)
```

# Removing Stop Words

```python
def remove_stop_words(text):
    words = text.split()
    return " ".join([w for w in words if not w in stops and not w in stop_word_comp and len(w) >= 2])
```

# Stemming

```python
def stem_text(text):
    words = text.split()
    words = [ARstemmer.stem(w) for w in words]
    return " ".join(words)
```

```python
def preprocess_text(text):
    text = normalize_arabic(text)
    text = remove_diacritics(text)
    text = remove_stop_words(text)
    text = stem_text(text)
    return text
```

# Load AIArabicBinary Dataset

```python
df = pd.read_csv("../data/processed/AIArabic_binary.csv")
```

# Clean AIArabicBinary Dataset

```python
df['clean_text'] = df['text'].astype(str).apply(preprocess_text)
```
